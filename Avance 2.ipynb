# Descripción de los datos

 - La información en la base de datos está representada por tablas.
 - Dado que tenemos diferentes tablas, explicaremos las columnas por secciones:

### Coverage modifications per group: Aquí manejaremos 4 columnas:

1. Grupo: Aquí se clasifican el total de asegurados en 5 grupos de acuerdo
con el tipo de seguro con el que cuentan.
2. Deducible: De acuerdo con el seguro con el que cuentan (grupo), es la
cantidad de deducible que pagara el asegurado. Ya hay un monto
prestablecido.
3. Coaseguro: De acuerdo con el seguro con el que cuentan (grupo), es la
cantidad de coaseguro que pagara el asegurado. Aquí se aplica un
porcentaje de acuerdo con el siniestro.
4. Límite de pago: Es la cantidad máxima que pagara la asegurada según el
grupo al que pertenece el asegurado.

### Diseases per person: Se encuentran 2 columnas: 

1. Subject: ID del asegurado.
2. Número de reclamaciones: Son las reclamaciones que ha realizado un asegurado es un periodo de un año.

### Payments 2015-2018: Para las tablas de Payments 2015 a Payments 2018, se usan las mismas 4 columnas:

1. Año: Hace referencia al año en el que se hiso la reclamación.
2. Tipo: Es el tipo de la enfermedad/accidente de la reclamación.
3. Grupo: A qué grupo pertenece el asegurado de acuerdo con la tabla “Coverage modifications per group”.
4. Pago: Es el monto que pago la aseguradora según el tipo de seguro.

**Conclusión de los estadísticos**

- El total de pagos por año por la aseguradora va desde los 12 hasta los 21 millones de pesos, no siendo uniforme la cantidad pagada año por año
- La varianza del año 2017 y 2018 es significativamente incrementada en comparación con los años 2015 y 2016, esto se ve reflejado en los máximos montos pagados en los años correspondientes.
- Como en el punto anterior, los máximos montos también desplazan a la media de los pagos por año.

- El mínimo monto pagado por los 4 años no parecen tener variabilidad significante.
